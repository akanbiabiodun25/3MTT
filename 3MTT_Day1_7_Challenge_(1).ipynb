```python
# Initialize the array size and number of operations
n = 5  # Size of the array
operations = [
    (1, 2, 100),
    (2, 5, 100),
    (3, 4, 100)
]
```

```python
# Initialize the array with zeros
arr = [0] * (n + 1)

# Apply the operations
for op in operations:
    start, end, value = op
    arr[start - 1] += value
    if end <= n:
        arr[end] -= value

# Calculate the maximum value after all operations
max_value = 0
current_value = 0
for i in arr:
    current_value += i
    if current_value > max_value:
        max_value = current_value

print("The maximum value in the array after all operations is:", max_value)
```

```python
df_cleaned['Title and Name'] = df_cleaned['Name'].str.split(',', expand=True)[1].str.split('(', expand=True)[0].str.strip()

# Display the updated DataFrame with the 'Title and Name' column
df_cleaned[['Name', 'Title and Name']].head()
```

```python
# Split the 'Name' column by the comma
df_cleaned[['Last Name', 'Title and Name']] = df_cleaned['Name'].str.split(',', expand=True)

# Extract the title and name without the parentheses content
df_cleaned['Title and Name'] = df_cleaned['Title and Name'].str.split('(', expand=True)[0].str.strip()
```

```python
# Drop the 'Name' column
df_cleaned = df_cleaned.drop(columns=['Name'])

# Merge the 'Title and Name' with 'Last Name' to create 'Cleaned_name'
df_cleaned['Cleaned_name'] = df_cleaned['Title and Name'] + ' ' + df_cleaned['Last Name']

# Optionally, drop the individual 'Title and Name' and 'Last Name' columns if no longer needed
df_cleaned = df_cleaned.drop(columns=['Title and Name', 'Last Name'])
```

```python
# Standardize the 'Ticket' column by removing extra spaces and converting to uppercase
df_cleaned['Ticket'] = df_cleaned['Ticket'].str.strip().str.upper()

# Optionally, remove any non-alphanumeric characters (e.g., special symbols) from the 'Ticket' column
df_cleaned['Ticket'] = df_cleaned['Ticket'].str.replace(r'[^A-Z0-9]', '', regex=True)

# If there is a pattern where you need to split tickets by a prefix (e.g., letters and numbers), you can do:
df_cleaned['Ticket_Prefix'] = df_cleaned['Ticket'].str.extract(r'([A-Z]+)', expand=False)  # Extract letter prefix
df_cleaned['Ticket_Number'] = df_cleaned['Ticket'].str.extract(r'(\d+)', expand=False)     # Extract number part

# Display the cleaned dataframe
df_cleaned[['Ticket', 'Ticket_Prefix', 'Ticket_Number']].head()
```

```python
# Create a new column 'Cleaned_Ticket' by removing any non-numeric characters
df_cleaned['Cleaned_Ticket'] = df_cleaned['Ticket'].str.extract(r'(\d+)', expand=False)

# Display the updated dataframe to check the 'Cleaned_Ticket' column
df_cleaned[['Ticket', 'Cleaned_Ticket']].head()
```

```python
# Create a new column 'Cleaned_Ticket' by extracting only the numeric part of the 'Ticket' column
df_cleaned['Cleaned_Ticket'] = df_cleaned['Ticket'].str.extract(r'(\d+)', expand=False)

# Drop the 'Ticket', 'Ticket_Prefix', and 'Ticket_Number' columns (if Ticket_Prefix and Ticket_Number were already created)
df_cleaned.drop(columns=['Ticket', 'Ticket_Prefix', 'Ticket_Number'], inplace=True, errors='ignore')

# Display the updated dataframe with only the 'Cleaned_Ticket' column
df_cleaned[['Cleaned_Ticket']].head()
```
